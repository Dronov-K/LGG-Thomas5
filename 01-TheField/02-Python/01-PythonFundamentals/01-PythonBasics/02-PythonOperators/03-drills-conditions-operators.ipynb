**Calculate the sum of all items and store it in a `order_price` variable**

```python
prices = {
    "milk": 0.45,
    "cider": 3.85,
    "flour": 0.9,
    "butter": 0.77,
    "nutella": 1.87
}

quantity = {
    "milk": 2,
    "cider": 3,
    "flour": 1,
    "butter": 1,
    "nutella": 1
}

order_price = sum(prices[item] * quantity[item] for item in prices)

print(order_price)
```

**Create a variable `allowance_money` which has a value of 20 and then create an algorithm that calculates the available money by subtracting the price of the order.**

If there is enough money, record the following sentence in the variable `message` and subtract the expense from `allowance_money` :
- message = "You have spent" + `order_price` + "you have left" + `allowance_money`

If there is not enough money, record the following sentence in the `message` variable:
- message= "Sorry you're missing *amount_missing* euros"

If there is 0 left, record the following sentence in the `message` variable:
- message = "You are broke!"


```python
def check_allowance(price):
    """Сhecks and withdraws the balance depending on the purchase price"""

    allowance_money = 20
    remaining_money = allowance_money - price

    if remaining_money > 0:
        message = f"You have spent {price} euros, you have left {remaining_money:.2f} euros."
    elif remaining_money == 0:
        message = "You are broke!"
    else:
        message = f"Sorry, you're missing {-remaining_money:.2f} euros."

    return message


if __name__ == '__main__':
    print(check_allowance(float(input())))
```

**10. Write a script that asks you to enter 2 strings and displays the largest of the 2 strings (the one with the most characters).**

```python
def get_larger_string(str_1: str, str_2: str) -> None:
    """Compares strings by number of characters and outputs the larger one"""
    print(max(str_1, str_2, key=len))


if __name__ == '__main__':
    get_larger_string(input(), input())
```

**11. Write a script that converts euros into dollars.**
- The program will start by asking the user to indicate with a character 'E' or '$' depending on the currency of the amount they are entering.
- Then the program will ask you to enter the amount and display the conversion.

```python
def convert_currency(currency: str, amount: int | float) -> str:
    convertion_rates = {'E': 1.1,
                        '$': 1 / 1.1}
    currency_to_convert = '$' if currency == 'E' else 'E'
    try:
        converted_amount = amount * convertion_rates[currency]
    except KeyError:
        return "Invalid currency symbol. Only works with Euros and Dollars. Please enter 'E' or '$'."

    return f'{amount} {currency} = {converted_amount:.2f} {currency_to_convert}'


if __name__ == '__main__':
    currency = input("Enter the currency symbol 'E' or '$': ")
    amount = float(input("Enter amount: "))

    print(convert_currency(currency, amount))
```
